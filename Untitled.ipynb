```python
for filter, filtered_data in expected_filtered_data.items():
    if filter in ["svd"]:
        cover_mask = torch.full((filtered_data.shape[0], 10), False, dtype=torch.bool)
        data_min = torch.min(filtered_data)
        data_max = torch.max(filtered_data)
        data_range = torch.max(filtered_data) - torch.min(filtered_data)
        print(data_min, data_max, data_range)
        # width of each interval in the cover
        cover_width = data_range / (10 - (10 - 1) * 0.3)
        print(cover_width)
        last = data_min + (10 - 1) * (1 - 0.3) * cover_width
        lows = torch.zeros(10)
        for i in range(10):
            lows[i] = (data_min) + (i) * (1 - 0.3) * cover_width
        highs = lows + cover_width
        # construct boolean cover
        for j, pt in enumerate(filtered_data):
            for i in range(10):
                # if j==0:
                #     print(i)
                #     print(pt > lows[i])
                #     print(torch.isclose(pt, lows[i]))
                #     print("AND")
                #     print(pt < highs[i])
                #     print(torch.isclose(pt, highs[i]))
                if (pt > lows[i] or torch.isclose(pt, lows[i])) and (
                    pt < highs[i] or torch.isclose(pt, highs[i])
                ):
                    cover_mask[j, i] = True
        # delete empty covers
        keep = torch.full([10], True, dtype=torch.bool)
        count_falses = 0
        for i in range(10):
            for j in range(filtered_data.shape[0]):
                if not cover_mask[j, i]:
                    count_falses += 1
            if count_falses == filtered_data.shape[0]:
                keep[i] = False
            count_falses = 0
        print(filter)
        print(torch.t(torch.t(cover_mask)[keep]))
        print(torch.hstack((lows.reshape([10, 1]), highs.reshape([10, 1]))))
        print("---------------")

resolution = 10
gain = 0.3
filtered_data = expected_filtered_data["svd"]


def get_cover(filtered_data, resolution, gain):
    data_min = torch.min(filtered_data)
    data_max = torch.max(filtered_data)
    data_range = torch.max(filtered_data) - torch.min(filtered_data)
    print(data_min, data_max, data_range)
    # width of each interval in the cover
    cover_width = data_range / (resolution - (resolution - 1) * gain)
    last_lower_endpoint = data_min + cover_width * (resolution - 1) * (1 - gain)
    lower_endpoints = torch.linspace(data_min, last_lower_endpoint, resolution)
    upper_endpoints = lower_endpoints + cover_width
    cover_intervals = torch.hstack(
        (
            lower_endpoints.reshape([resolution, 1]),
            upper_endpoints.reshape([resolution, 1]),
        )
    )
    # want a n x resolution Boolean tensor
    lower_values = torch.gt(filtered_data, lower_endpoints)
    upper_values = torch.lt(filtered_data, upper_endpoints)
    # need to check close values to deal with some endpoint issues
    lower_is_close_values = torch.isclose(filtered_data, lower_endpoints, atol=1e-3)
    upper_is_close_values = torch.isclose(filtered_data, upper_endpoints, atol=1e-3)
    # construct the boolean mask
    mask = torch.logical_and(
        torch.logical_or(lower_values, lower_is_close_values),
        torch.logical_or(upper_values, upper_is_close_values),
    )
    # remove empty intervals from cover
    non_empty_covers = torch.any(mask, 0)
    print("point", filtered_data[0], "coverwidth", cover_width)
    print(cover_intervals)
    print(lower_values[0, 9], lower_is_close_values[0, 9])
    print("AND")
    print(upper_values[0, 9], upper_is_close_values[0, 9])
    return mask[:, non_empty_covers]
```
